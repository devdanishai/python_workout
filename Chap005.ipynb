### summary


1. **`if / else` above the `for` loop** → Check whether a list is empty.
2. **`in` and `not in` operators** → Check whether a value exists in a list.
3. **Comparing two lists** → Use a `for` loop with `in`/`not in` to compare items.
4. **Multiple `if` statements** → Check multiple independent conditions (every `if` is evaluated).
5. **`if / elif / else`** → Choose **one** outcome from multiple conditions.

The fifth point is also an important takeaway from Chapter 5:

* Multiple `if` statements = all conditions can run.
* `if`/`elif`/`else` = only the first matching condition runs.
