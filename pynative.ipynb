 Write a program to check if a given number is a palindrome (reads the same forwards and backwards).

```python
# Case 1: number = 121
# Case 2: number = 125
def r_check(num):
  num = str(num)
  r_num = num[::-1]
  if num == r_num:
    return True
  else:
    return False

number = 121
call = r_check(number)
print(call)
```

```python
def r_check(num):
    num = str(num)
    r_num = num[::-1]

    if num == r_num:
        return True
    else:
        return False

print(r_check(121))  # True
print(r_check(125))  # False
```

Merging Lists with Parity Filtering

```python
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]
new_list=[]
for i in list1:
  if i % 2 != 0:
    new_list.append(i)

for i in list2:
  if i % 2 == 0:
    new_list.append(i)

print(new_list)
```
